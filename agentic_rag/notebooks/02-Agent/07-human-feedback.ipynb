```python
from langsmith import Client
from dotenv import load_dotenv
```

```python
load_dotenv()
ls_client = Client()
```

### Attach feedback to a specific trace in LangSmith

019f2bc1-c6bb-7310-b45e-044f0245f1f4

```python
RUN_ID = "019f2bc1-c6bb-7310-b45e-044f0245f1f4"
```

```python
ls_client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
ls_client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by hyh",
    feedback_source_type="api"
)
```

```python
ls_client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by hyh",
    score=0,
    feedback_source_type="api"
)
```
